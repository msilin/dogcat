```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("DF_cities.csv", index_col=[0])
```

```python
#Favorite Image From Individual Users
df_total_ind = df.groupby(["IP","cookie","path"]).size().reset_index(name='count')
df3 = df_total_ind.groupby(["IP","cookie"]).max().reset_index()
df3 = df3.groupby('path').size().sort_values(ascending=False)
df3
```

```python
df_total_ind.groupby(["IP","cookie"]).max().reset_index()
```

```python
most_liked = df3.plot.bar(figsize=(25,15),fontsize=20,width=.9,title='Location')
most_liked.set_xlabel("Image", fontsize=26)
most_liked.set_ylabel("Individual Users Favorite", fontsize=26)
most_liked.title.set_size(35)
most_liked.get_figure().savefig('most_liked.png')
```

```python
#Local Time Zones
from datetime import datetime
              
df_t = df.copy(deep=True)
df_t['datetime'] = df_t[['date','time']].apply(lambda x: datetime.strptime(''.join(x)[:-1],'%m/%d/%Y %H:%M:%S'), axis=1).dt.tz_localize('UTC')

df_t = df_t.loc[df_t['CountryCode'].notna()]
```

```python
from pytz import country_timezones 
df_convert = df_t
df_convert['datetime'] = df_convert.apply(lambda x: x['datetime'].tz_convert(country_timezones(x['CountryCode'])[0]), axis=1)
```

```python
df2 = df_convert.copy(deep=True)
df2['localtime']=df_convert['datetime'].astype(str).str[10:19]
df2 = df2.drop(['time','datetime'], 1)
df2.to_csv("localtime.csv",index=False)
```

```python
df3 = df2
df3['localtime']=df3['localtime'].str[:3]
df3 = df3.groupby(['localtime']).size().sort_values(ascending=False)
```

```python
ax = df3.plot.bar(figsize=(25,15),fontsize=20,width=.9,title='Local Time ')
ax.set_ylabel("Clicks", fontsize=26)
ax.set_xlabel("Times", fontsize=26)
ax.title.set_size(35)
ax.get_figure().savefig('Local_Times.png')
```
