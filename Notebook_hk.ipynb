```python
df3 = df

df3 = df3.sort_values(by=['IP'])
Preferences = df3[500:7000].groupby(['IP', 'path']).size().reset_index(name='value')
Preferences
```

```python
pivot_df = Preferences.pivot(index='IP', columns='path', values='value')
pivot_df
```

```python
colors = ["#eb1010","#e84a4a","#eb7a7a","#85a4ed", "#4e7eed","#1556ed"]
pivot_df.loc[:,['kitten1','kitten2', 'kitten3', 'puppy1', 'puppy2', 'puppy3']].plot.bar(stacked=True, color=colors, figsize=(10,7))
```

```python
cp = df3[df3.CountryCode != 'US']

sort = cp.groupby(['CountryName']).size().sort_values(ascending=False).index.tolist()
CountryPreferences = cp.groupby(['CountryName', 'path']).size().reset_index(name='value')
pivot_dfcp = CountryPreferences.pivot(index='CountryName', columns='path', values='value')
colors = ["#eb1010","#e84a4a","#eb7a7a","#85a4ed", "#4e7eed","#1556ed"]
graph = pivot_dfcp.loc[:,['kitten1','kitten2', 'kitten3', 'puppy1', 'puppy2', 'puppy3']].reindex(sort).plot.bar(stacked=True, color=colors, figsize=(15,7))
graph.get_figure().savefig('preferences_by_country_withUS.png')
```

```python
import pycountry_convert as pc

df9 = pd.read_csv("DF_cities.csv")

def country_to_continent(country_name):
    country_alpha2 = pc.country_name_to_country_alpha2(country_name)
    country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
    country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    return country_continent_name

# df9.at[0, 'CountryName']

countries = df9[df9.CountryName.notnull()]

for i in countries.itertuples():
    countries.at[i.Index, 'CountryName'] = country_to_continent(i.CountryName)
    
countries
```
